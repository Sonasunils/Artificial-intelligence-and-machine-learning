# NumPy Array Creation
This notebook demonstrates different ways to create arrays using NumPy.

```python
import numpy as np
```

## Create array from Python list

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr1
```

## Create 2D array

```python
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
arr2
```

## Zeros array

```python
zeros_arr = np.zeros((3, 4)) # 3 column and 4 row
zeros_arr
```

## Ones array

```python
ones_arr = np.ones((2, 3))
ones_arr
```

## Range array

```python
range_arr = np.arange(1, 10, 2) #start step stop

range_arr
```

## Linspace array

```python
linspace_arr = np.linspace(0, 1, 5)
linspace_arr
```

## Identity matrix

```python
identity_arr = np.eye(4)
identity_arr
```

## Random array

```python
random_arr = np.random.rand(3, 3)  #arrange with random values of 0 to 1
random_arr
```
