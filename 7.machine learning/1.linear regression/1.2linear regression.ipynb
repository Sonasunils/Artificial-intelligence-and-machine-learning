```python
# Simple Linear Regression: House Price Prediction
import numpy as np
from sklearn.linear_model import LinearRegression

# Training Data
# House Size in sq ft (X)
X = np.array([[500], [800], [1000], [1200], [1500], [2000]])

# House Price in Rupees (Y)
# Using approximate linear pattern: Price = 2500×Size + 150000
Y = np.array([2500*x + 150000 for x in X]).reshape(-1, 1)

# Create the model
model = LinearRegression()

# Train the model
model.fit(X, Y)

# Print the learned slope (m) and intercept (c)
print("Slope (m):", model.coef_[0][0])
print("Intercept (c):", model.intercept_[0])

# Predict price for a new house size
new_size = np.array([[1800]])  # 1800 sq. ft.
predicted_price = model.predict(new_size)

print("Predicted price for 1800 sq ft:", predicted_price[0][0])
```

```python
# Simple Linear Regression: Salary Prediction
import numpy as np
from sklearn.linear_model import LinearRegression

# Training Data
# X = Years of Experience (independent variable)
X = np.array([[1], [2], [3], [4], [5], [6]])

# Y = Salary (dependent variable) generated using:
# Salary = 5000 × Experience + 20000
Y = np.array([5000*x + 20000 for x in X]).reshape(-1, 1)

# Create the model
model = LinearRegression()

# Train the model
model.fit(X, Y)

# Print model parameters
print("Slope (m):", model.coef_[0][0])
print("Intercept (c):", model.intercept_[0])

# Predict salary for 7 years of experience
new_exp = np.array([[7]])
predicted_salary = model.predict(new_exp)

print("Predicted salary for 7 years experience:", predicted_salary[0][0])
```
