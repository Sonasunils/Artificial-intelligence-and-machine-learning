```python
# Sales Data Analysis Project

# Description:
# Analyze monthly sales data and visualize trends.

# Uses

# Pandas: CSV reading, column operations

# NumPy: total and average calculations

# Matplotlib: line chart

# Concepts

# CSV handling

# Data aggregation

# Trend visualization


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Read CSV file
data = pd.read_csv("sales_data.csv")

# Display data
print("Sales Data:")
print(data)

# Convert Sales column to NumPy array
sales_array = np.array(data["Sales"])

# Calculate total and average sales
total_sales = np.sum(sales_array)
average_sales = np.mean(sales_array)

print("\nTotal Sales:", total_sales)
print("Average Monthly Sales:", average_sales)

# Plot sales trend
plt.figure()
plt.plot(data["Month"], data["Sales"], marker='o')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Trend")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
