```python
#2. Read specific columns using read_csv

df = pd.read_csv("people_data.csv", usecols=["First Name", "Email"])
print(df)
```

```python
#3.Setting an index column

df = pd.read_csv("people_data.csv", index_col="First Name")
print(df)
```

```python
# 3. Handling Missing Values Using read_csv

df = pd.read_csv("Book1.csv", na_values=["na", "Unknown"])
df
```

```python
#4. Reading CSV Files with Different Delimiters
#create a new file named sample.csv

import pandas as pd

# Sample data stored in a multi-line string
data = """totalbill_tip, sex:smoker, day_time, size
16.99, 1.01:Female|No, Sun, Dinner, 2
10.34, 1.66, Male, No|Sun:Dinner, 3
21.01:3.5_Male, No:Sun, Dinner, 3
23.68, 3.31, Male|No, Sun_Dinner, 2
24.59:3.61, Female_No, Sun, Dinner, 4
25.29, 4.71|Male, No:Sun, Dinner, 4"""

# Save the data to a CSV file
with open("sample.csv", "w") as file:
    file.write(data)
print(data)


```

```python
#5.Reading 3 rows

df = pd.read_csv('sample.csv', nrows=3)
df
```

```python
#6. Using skiprows in read_csv()

df= pd.read_csv("people_data.csv")
print("Previous Dataset: ")
print(df)
# using skiprows
df = pd.read_csv("people_data.csv", skiprows = [1,3])
print("Dataset After skipping rows: ")
print(df)
```

```python
#7. Parsing Dates (parse_dates)

df = pd.read_csv("people_data.csv", parse_dates=["Date of birth"])
print(df.info())
```
