```python
#concatanating two excel records
import pandas as pd

file1 = 'marksheet_1.xlsx'
file2 = 'marksheet_2.xlsx'

sheet1 = pd.read_excel(file1, index_col=0)
sheet2 = pd.read_excel(file2, index_col=0)

# Combine both Excel files
newData = pd.concat([sheet1, sheet2])
print(newData)
```

```python
#Head() and Tail() methods in Pandas
#Return 5 rows
print(newData.head())
print(newData.tail())
```

```python
#Return the shape 
newData.shape
```

```python
#Sort the values

sorted_column = newData.sort_values(['English'], ascending = False)
sorted_column
```

```python
#We can get the statistical information like mean, max, min, etc. about the data frame using the describe() method

newData.describe()
```

```python
#we cam also find seprate means

newData['English'].mean()
```

```python
# dinding the toatal marks and the head return dirst 5 columns 

newData['Total Marks'] = newData["English"] + newData["Maths"] + newData["Science"]
newData['Total Marks'].head()
```

```python
# we can export the data back to an Excel file using the method to_excel

newData.to_excel('Output File.xlsx')
```

```python
#Module 5 

# reading text files with pandas

#Using read_csv()
# Using read_table()
# Using read_fwf()  These methods can read text files in pandas

# Read Text Files with Pandas using read_csv()

# importing pandas
import pandas as pd

# read text file into pandas DataFrame
df = pd.read_csv("players_list.txt", sep=" ")

# display DataFrame
print(df)
```

```python

import pandas as pd

# read text file into pandas DataFrame and
# create header
df = pd.read_csv("players_list.txt", sep=" ", header=None)

# display DataFrame
print(df)
```

```python


# read text file into pandas DataFrame and create 
# header with names
df = pd.read_csv("players_list.txt", sep=" ", header=None, 
                 names=["Team1", "Team2"])

# display DataFrame
print(df)
```

```python

#using read table method

# read text file into pandas DataFrame
df = pd.read_table("players_list.txt", delimiter=" ")

# display DataFrame
print(df)
```

```python


# Read Text Files with Pandas using read_fwf()

# importing pandas
import pandas as pd

# read text file into pandas DataFrame
df = pd.read_fwf("players_list.txt")

# display DataFrame
print(df)
```
