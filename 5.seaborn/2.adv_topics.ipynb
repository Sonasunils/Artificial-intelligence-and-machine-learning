```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load a dataset
data = sns.load_dataset("iris")

# Compute correlation matrix
corr = data.corr(numeric_only=True)

# Create heatmap
sns.heatmap(corr, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title("Correlation Heatmap - Iris Dataset")
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("iris")

sns.pairplot(data, hue="species", palette="husl")
plt.suptitle("Pair Plot - Iris Dataset", y=1.02)
plt.show()
```

```python
#customizing pair plot

sns.pairplot(data, 
             kind="kde",  # or 'reg' for regression lines
             hue="species", 
             diag_kind="kde",
             markers=["o", "s", "D"])
plt.show()
```

```python
#face grid

import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset("tips")

g = sns.FacetGrid(data, col="sex")
g.map(sns.histplot, "total_bill")

plt.show()
```

```python
#face grid with rows and columns

g = sns.FacetGrid(data, row="smoker", col="time", margin_titles=True)
g.map(sns.scatterplot, "total_bill", "tip", alpha=0.7)
plt.show()
```

```python
# with customized data

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Create a custom dataset
data = {
    'Math': [85, 78, 90, 66, 95],
    'Science': [80, 70, 88, 60, 92],
    'English': [75, 82, 89, 72, 85],
    'Computer': [92, 85, 94, 78, 98]
}

students = pd.DataFrame(data, index=['A', 'B', 'C', 'D', 'E'])

# Step 2: Find correlation
corr = students.corr()

# Step 3: Plot a customized heatmap
plt.figure(figsize=(7,5))
sns.heatmap(corr, annot=True, cmap='magma', linewidths=1, linecolor='white', square=True)
plt.title('Subject Correlation Heatmap', fontsize=14, fontweight='bold')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Create custom dataset
data = {
    'Age': [21, 22, 23, 24, 25, 26, 27, 28],
    'Experience': [1, 2, 1, 3, 4, 3, 5, 6],
    'Salary': [22000, 25000, 24000, 28000, 30000, 32000, 35000, 38000],
    'Department': ['IT', 'IT', 'HR', 'HR', 'Finance', 'Finance', 'IT', 'HR']
}

employees = pd.DataFrame(data)

# Step 2: Create a pair plot
sns.pairplot(employees, hue='Department', diag_kind='kde', palette='husl', markers=['o','s','D'])
plt.suptitle("Employee Data Pair Plot", y=1.02)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Step 1: Custom dataset for product sales
data = {
    'Month': ['Jan', 'Jan', 'Feb', 'Feb', 'Mar', 'Mar', 'Apr', 'Apr', 'May', 'May'],
    'Product': ['A', 'B'] * 5,
    'Sales': [200, 150, 220, 170, 250, 180, 270, 190, 300, 210],
    'Region': ['North', 'South', 'North', 'South', 'North', 'South', 'North', 'South', 'North', 'South']
}

sales = pd.DataFrame(data)

# Step 2: Create a FacetGrid
g = sns.FacetGrid(sales, col='Region', hue='Product', height=4, aspect=1)
g.map(sns.barplot, 'Month', 'Sales', order=['Jan', 'Feb', 'Mar', 'Apr', 'May'], dodge=True)
g.add_legend()
g.set_titles("Region: {col_name}")
plt.suptitle("Monthly Sales Comparison by Region and Product", y=1.05)
plt.show()
```
