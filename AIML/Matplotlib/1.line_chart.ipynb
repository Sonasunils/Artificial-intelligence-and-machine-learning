```python
# Example 1: In this example, a simple line chart is generated using NumPy to define data values.
# The x-values are evenly spaced points, and the y-values are calculated as twice the corresponding x-values.


import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])  # X-axis 
y = x*2  # Y-axis

plt.plot(x, y)  
plt.show()
```

```python
# Example 2: Since labeling is necessary for understanding the chart dimensions. 

import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x*2

plt.plot(x, y)
plt.xlabel("X-axis")        # Label for the X-axis
plt.ylabel("Y-axis")        # Label for the Y-axis
plt.title("sona's graph")  # Chart title
plt.show()
```

```python
#Chart annotations 

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(8, 6))
plt.plot(x, y, marker='o', linestyle='-')

# Add annotations
# for i, (xi, yi) in enumerate(zip(x, y)):
#     plt.annotate(f'({xi}, {yi})', (xi, yi))

for i, (xi, yi) in enumerate(zip(x, y)):
    plt.annotate(f'({xi}, {yi})', (xi, yi))

plt.title('Line Chart with Annotations')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

plt.grid(True)
plt.show()
```

```python
#Example 4 multiline graph using matplotlib

import matplotlib.pyplot as plt
import numpy as np


x = np.array([1, 2, 3, 4])
y = x*2

plt.plot(x, y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Any suitable title")
plt.show()  # show first chart

# plt.figure()
x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]
plt.plot(x1, y1, '-.')
plt.show()
```

```python
#importance of plt.figure()
# example 5 multi plot on the same axis


import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x * 2

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

# 👇 Plot both lines without calling plt.figure()
plt.plot(x, y, label='Line 1')
# plt.figure()
plt.plot(x1, y1, '-.', label='Line 2')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Both Lines in One Chart")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x*2

plt.plot(x, y)

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

plt.plot(x, y1, '-.')
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')

plt.fill_between(x, y, y1, color='green', alpha=.5)
plt.show()
```
