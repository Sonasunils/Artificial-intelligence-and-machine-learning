```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
data = pd.DataFrame({
    "genre": ["Pop", "Pop", "Rock", "Rock", "Jazz", "Jazz"],
    "gender": ["Male", "Female", "Male", "Female", "Male", "Female"],
    "count": [50, 60, 40, 55, 30, 45]
})

# Line plot with hue
sns.lineplot(x="genre", y="count", data=data, hue="gender")
sns.set(style="darkgrid")

plt.title("Music Genre Popularity by Gender")
plt.show()



```
